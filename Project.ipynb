# *Cross-Border Commuters in Switzerland*

This work explores the per qauerter cross-border commuters count over years, since 1999. It has been on a rise for a while and it's an interesting topic in local politics (https://lenews.ch/2017/02/24/nearly-320000-cross-border-workers-in-switzerland/).

Source: https://www.pxweb.bfs.admin.ch/pxweb/en/px-x-0302010000_108/-/px-x-0302010000_108.px/table/tableViewLayout2/?rxid=e43cc8c3-cf50-4234-b0fa-24573f06e733

Notes (copied from the source):

*"The cross-border commuters Statistics was revised in the fourth quarter of 2015.
    The cross-border commuters Statistics is based on the data from the Central Information Service on Migration (ZEMIS), the data from the old-age insurance system (AHV), the database of the Vocational Education and Training (SBG) and, for the time series before the fourth quarter 2010, the Job Statistics (BESTA).
    Data users need to bear in mind the following: although the Central Migration Information System (ZEMIS) is the main source for the breakdowns disseminated, the results are subject to a certain degree of uncertainty. The coding of some variables is neither mandatory nor subject to plausibility tests at source. Furthermore, certain job changes are not necessarily reported and are only recorded in the register when cross-border commuter permits are renewed, possibly affecting the place of work, economic activity, and to a lesser degree, employment status.
    **Finally, during the compilation of the statistics, the adjustments made and retropolation or extrapolation produce results with decimal digits even though the statistics measure persons.
    The figures are rounded off to one decimal place. Subsequently, it is possible that the total of the breakdowns does not exactly equal the sum of the variables.** The figures with all decimal places are available in the csv file.
    As of 1 June 2007, nationals of the EU-17/EFTA Member States are no longer bound to border zones. Cross-border commuters can therefore seek employment in all cantons."*

## Predictions

### Prediction per Year

Let's fit the yearly averages to a linear regression, except for the last few values that will be used for testing.

```python
from sklearn.linear_model import LinearRegression

# Linear regression
lr = LinearRegression()
data_y = data.iloc[0][-20:-4].values # train set, all but the last 3 years
lr.fit(np.arange(1999, 2015).reshape(-1,1), data_y) # Fit to train data
```

```python
plt.figure(figsize=(15,5))

plt.plot(range(1999, 2018), data.iloc[0][82:-1].values, 'bo-', label='Historical data', zorder=4)
plt.plot(range(2017, 2018+1), data.iloc[0][-2:].values, 'ro-', label='2018 (Q1 - Q3)', zorder=3)
plt.plot(range(1999, 2016), lr.predict(np.array(range(1999, 2016)).reshape(-1,1)), 'g--', label='Regression line')
plt.plot(range(2015, 2021), lr.predict(np.array(range(2015, 2021)).reshape(-1,1)), 'go--', mfc='none',
        label='Predicted values')
plt.xticks(range(1999, 2021), range(1999, 2021), rotation=45)
plt.legend()
plt.show()
```

The blue points represent the historical data, avereged per year. Green points are predictions, three last years where all data is available, but was used as a test set (2015-2017) and future points (2018-2020, since incomplete data is available for 2018). The year 2018 is shown in red, since the trend could still change once that Q4 data is available. 

It can be noted that the regression line approximates the values fairly and the decrease in 2018 (if it happens, once that total data is available) might be temporary. 

```python
# Compare models with mean absolute error (MAE)
def MAE(y, y_pred):
    return np.mean(np.abs(y - y_pred))
```

```python
mae_lr = MAE(data.iloc[0][-4:-1].values, lr.predict(data.iloc[0][-4:-1].index.values.reshape(-1,1)))
mae_lr
```

```python
from sklearn.dummy import DummyRegressor

# Create the DummyRegressor object
dummy = DummyRegressor(strategy='mean')
dummy.fit(np.arange(1999, 2015).reshape(-1,1), data_y)
```

```python
mae_dummy = MAE(data.iloc[0][-4:-1].values, dummy.predict(data.iloc[0][-4:-1].index.values.reshape(-1,1)))
mae_dummy
```

```python
r2_lr = lr.score(np.arange(2015, 2018).reshape(-1,1), data[np.arange(2015, 2018)].loc[0].values)
r2_lr
```

```python
dummy.score(np.arange(2015, 2018).reshape(-1,1), data[np.arange(2015, 2018)].loc[0].values)
```

The MAE score is rather low, comparing to the number of commuters and the baseline (MAE). However, the R^2 factor is less than 0. That is because the regression line is completely below the test set, so that sum of squared differences between the atual values and their mean is smaller than the sum of squared differences betwen actual values and predictions. Let's compute R^2 manually to clarify:

```python
y = [data.loc[0][i] for i in range(2015, 2018)]
y_m = np.mean(y)
y_pred = lr.predict(np.array(range(2015, 2018)).reshape(-1,1))

ss_tot = np.sum(np.square(y-y_m))
print('SS_tot: ', ss_tot)

ss_res = np.sum(np.square(y-y_pred))
print('SS_res: ', ss_res)

print('R^2: ', 1 - ss_res / ss_tot)
```

### Prediction per Year and Quarter

Since there is a noticabletrend in increase of commuters from Q1 to Q2 and, a polynomial dependency of commuters to quarter should be observed. The idea is to keep the dependency to a year linear and to experiment with different polynomials of quarter, up to the fourth degree. A special table will be created for that purpose, containing the year, qurater, the 2nd, the 3rd, and the 4th powers of quarters. Note: a polynomial dependency could have been created automatically by the library call to polynomial features, but in that case the polynomial features of year would be created, too, which wouldn't make much sense.

```python
data_q4 = pd.DataFrame(columns=['Year', 'Q', 'Q^2', 'Q^3', 'Q^4', 'Commuters'])
```

```python
# Populate the table with years, quarter, qurater^2, quarter^3,, quarter`4,
# and the number of commuters per year and qaurter

years = range(1999, 2018+1)
index = 0
for year in (years):
    for q in range(1,4+1):
        value = data.loc[0][str(year) + 'Q' + str(q)]
        data_q4.loc[index] = [year, q, q**2, q**3, q**4, value]
        index += 1
        if(year == 2018 and q == 3):
            break
```

```python
data_q4.head()
```

```python
data_q4.shape
```

```python
from sklearn.model_selection import train_test_split

# split 79 values into 60 train and 19 test values
X_train, X_test, y_train, y_test = train_test_split(data_q4[['Year', 'Q', 'Q^2', 'Q^3', 'Q^4']], data_q4.Commuters,
    test_size=19, random_state=0
)
```

```python
lr_yq1 = LinearRegression()
lr_yq1.fit(X_train[['Year', 'Q']], y_train)

dummy_yq1 = DummyRegressor()
dummy_yq1.fit(X_train[['Year', 'Q']], y_train)

lr_yq1.coef_
```

```python
lr_yq2 = LinearRegression()
lr_yq2.fit(X_train[['Year', 'Q', 'Q^2']], y_train)

dummy_yq2 = DummyRegressor()
dummy_yq2.fit(X_train[['Year', 'Q', 'Q^2']], y_train)

lr_yq2.coef_
```

```python
lr_yq3 = LinearRegression()
lr_yq3.fit(X_train[['Year', 'Q', 'Q^2', 'Q^3']], y_train)

dummy_yq3 = DummyRegressor()
dummy_yq3.fit(X_train[['Year', 'Q', 'Q^2', 'Q^3']], y_train)

lr_yq3.coef_
```

```python
lr_yq4 = LinearRegression()
lr_yq4.fit(X_train[['Year', 'Q', 'Q^2', 'Q^3', 'Q^4']], y_train)

dummy_yq4 = DummyRegressor()
dummy_yq4.fit(X_train[['Year', 'Q', 'Q^2', 'Q^3', 'Q^4']], y_train)

lr_yq4.coef_
```

As expected, the contribution of a year doesn't change much, but the quarter coefficients change a lot as new powers are added.

```python
mae_q1 = MAE(lr_yq1.predict(X_test[['Year', 'Q']]), y_test)
```

```python
MAE(dummy_yq3.predict(X_test[['Year', 'Q', 'Q^2', 'Q^3']]), y_test)
```

```python
results = pd.DataFrame(columns=['Variables', 'MAE', 'R^2'])
results.loc[0] = ['Yearly average', mae_lr, r2_lr]
results.loc[1] = ['Year and Q', mae_q1, r2_q1]
results.loc[2] = ['Year, Q, and Q^2', mae_q2, r2_q2]
results.loc[3] = ['Year, Q, Q^2, and Q^3', mae_q3, r2_q3]
results.loc[4] = ['Year, Q, Q^2, Q^3, and Q^4', mae_q4, r2_q4]
results
```

It is evident that using quarter benefits to prediction, both in terms of MAE and R^2. The best result in terms of MAE is achieved for linear regression in year and quarter only. In terms of R^2 there is a minor benefit of using higher powers of quarter, however, but at the cost of higher MAE and more complex model. It is also interesting to notice that the models with the 3rd degree and the 4th degree polynomial in Q yield almost the same evaulation results! However, printing more decimal places reveals the difference.

```python
print('MAE_Q3: ', mae_q3)
print('MAE_Q4: ', mae_q4)
print('R^2_Q3: ', r2_q3)
print('R^2_Q4: ', r2_q4)
```

So, the best model is __linear regression in year and quarter__.
